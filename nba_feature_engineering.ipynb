# Define Target Variable

```
y=df['target_5yrs']
X=df.drop(columns=['target_5yrs'])
print(y.shape)
print(X.shape)
```

# Remove Non-Predictive Features

```
df=df.drop(columns=['Unnamed: 0','name'])
df.head()
```

# Missing Values

```
df.isnull().sum()
```

```
df.fillna(df.median(numeric_only=True), inplace=True)
```

# Correlation Analysis

```
corr=df.corr()
corr
```

```
upper=corr.where(np.triu(np.ones(corr.shape),k=1).astype(bool))
upper.head()
```

```
high_corr=[c for c in upper.columns if any(upper[c]>0.90)]
high_corr
```

```
reduced_df=df.drop(columns=high_corr)
reduced_df.head()
```

# Feature Engineering

```
reduced_df['pts_per_min']=reduced_df['pts']/reduced_df['min']
```

```
reduced_df['efficiency']=reduced_df['pts']+reduced_df['reb']+reduced_df['ast']+reduced_df['stl']+reduced_df['blk']-reduced_df['tov']
```
